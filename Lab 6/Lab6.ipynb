Для лабораторної роботи 6, яка стосується використання TensorFlow для генерації наборів даних та виконання операцій CRUD (створення, читання, оновлення, видалення) над даними, а також їх графічного аналізу з використанням Matplotlib, ось покрокова інструкція з кодом для створення TensorFlow тензорів, генерування наборів даних для регресії та класифікації, а також їх графічного аналізу.

Структура лабораторної роботи:
Імпорт необхідних бібліотек.
Створення та маніпулювання тензорами (скаляри, вектори, матриці).
Генерація наборів даних для регресії та класифікації.
Графічний аналіз наборів даних за допомогою Matplotlib.
Збереження результатів у GitHub.

```python
# Імпортуємо необхідні бібліотеки
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import sys
print(sys.executable)

# 1. Створення тензорів (скаляр, вектор, матриця)
scalar = tf.constant(5)  # Скаляр
print("Скаляр:")
print(scalar)

vector = tf.constant([1, 2, 3, 4, 5])  # Вектор
print("\nВектор:")
print(vector)

matrix = tf.constant([[1, 2], [3, 4], [5, 6]])  # Матриця
print("\nМатриця:")
print(matrix)

# 2. CRUD операції над даними:
# Створення тензора
tensor = tf.random.normal([3, 3])  # Тензор 3x3 з випадковими значеннями
print("\nВипадковий тензор 3x3:")
print(tensor)

# Читання тензора
tensor_value = tensor.numpy()  # Конвертуємо тензор у numpy масив для зручності
print("\nЗначення тензора як numpy масив:")
print(tensor_value)

# Оновлення тензора (можна змінити його значення, якщо потрібно)
tensor = tf.ones([3, 3])  # Оновлюємо тензор на масив одиничок
print("\nОновлений тензор (масив одиничок):")
print(tensor)

# Видалення тензора (в Python це не потрібно робити вручну, він буде зібраний сміттєзбирачем)
# tf.delete(tensor) # такої функції не існує, але це можна зробити в Python за допомогою del або del().

# 3. Генерація набору даних для регресії
# Створимо синтетичний набір даних для регресії (лінійна залежність)
X_reg = np.linspace(0, 10, 100)  # Ознака
y_reg = 2.5 * X_reg + 3 + np.random.randn(100)  # Цільова змінна з випадковим шумом

# Візуалізація набору даних для регресії
plt.figure(figsize=(8, 6))
plt.scatter(X_reg, y_reg, color='blue', label='Дані')
plt.plot(X_reg, 2.5 * X_reg + 3, color='red', label='Лінія регресії')
plt.title("Набір даних для регресії")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

# 4. Генерація набору даних для класифікації
# Створимо синтетичний набір даних для класифікації (дві групи)
from sklearn.datasets import make_classification

# Генерація набору даних для класифікації
X_class, y_class = make_classification(
    n_samples=100,        # кількість зразків
    n_features=2,         # кількість ознак
    n_classes=2,          # кількість класів
    n_clusters_per_class=1, # кількість кластерів на клас
    n_informative=2,      # кількість інформативних ознак
    n_redundant=0,        # кількість надлишкових ознак
    n_repeated=0          # кількість повторюваних ознак
)

# Візуалізація набору даних для класифікації
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(X_class[:, 0], X_class[:, 1], c=y_class, cmap=plt.cm.Paired)
plt.title('Synthetic Classification Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

import pandas as pd

# Збереження для регресії
df_reg = pd.DataFrame({'X': X_reg, 'y': y_reg})
df_reg.to_csv('regression_data.csv', index=False)

# Збереження для класифікації
df_class = pd.DataFrame(X_class, columns=['Feature1', 'Feature2'])
df_class['Label'] = y_class
df_class.to_csv('classification_data.csv', index=False)

print("\nДані збережено у файли 'regression_data.csv' та 'classification_data.csv'.")
```

Використовуємо TensorFlow для створення скаляра, вектора та матриці.
Далі здійснюємо CRUD операції, створюючи тензор випадкових значень, зчитуючи його, оновлюючи значення та "видаляючи" (або просто не використовуючи).
Генерація наборів даних для регресії:

Генеруємо лінійну залежність між ознакою X_reg та цільовою змінною y_reg з додаванням випадкового шуму для реалістичності.
Побудова графіка розсіювання для візуалізації цього набору даних.
Генерація наборів даних для класифікації:

Використовуємо бібліотеку sklearn для створення двокласового набору даних (по 2 ознаки).
Візуалізуємо дані на графіку розсіювання, де кольори вказують на приналежність до класу.
Графічний аналіз даних:

Використовуємо бібліотеку Matplotlib для побудови графіків розсіювання для регресії та класифікації, що дозволяє візуалізувати структуру даних.
