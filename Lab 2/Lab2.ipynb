```python
import pandas as pd
import numpy as np

random_data = np.random.randint(1, 10, (5, 3))
df_random = pd.DataFrame(random_data, columns=['Col1', 'Col2', 'Col3'])

data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 40, 45],
    'Salary': [50000, 60000, 70000, 80000, 90000]
}
df_dict = pd.DataFrame(data_dict)

print("DataFrame з масиву NumPy:\n", df_random.head(), "\n")
print("DataFrame зі словника:\n", df_dict.head(), "\n")

print("Статистика для DataFrame з масиву NumPy:\n", df_random.describe(), "\n")
print("Статистика для DataFrame зі словника:\n", df_dict.describe(), "\n")

print("Перший рядок з DataFrame (використання iloc):\n", df_random.iloc[0], "\n")
print("Інформація про Bob з DataFrame зі словника (використання loc):\n", df_dict.loc[df_dict['Name'] == 'Bob'], "\n")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error

x = np.linspace(0, 10, 100)  
y_true = 2 * x + 1 

noise = np.random.normal(0, 1, size=x.shape)
y_observed = y_true + noise 

plt.figure(figsize=(10, 6))
plt.plot(x, y_true, label='True function (y=2x+1)', color='blue')
plt.scatter(x, y_observed, label='Observed data (with noise)', color='red', s=10)
plt.title('Linear Function with Measurement Errors')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

mae = mean_absolute_error(y_true, y_observed)  
mse = mean_squared_error(y_true, y_observed)   

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")

results = {
    'Metric': ['MAE', 'MSE'],
    'Value': [mae, mse]
}
df_results = pd.DataFrame(results)
df_results.to_csv('error_metrics.csv', index=False)

print("Результати записані у файл 'error_metrics.csv'.")
```
